```python
import matplotlib.pyplot as plt

# Veriler
labels = ['Beğenilen Hesap Sayısı', 'Toplam Yorum Sayısı', 'Reels Yorum Sayısı', 'back like rate', 'response rate']
mgv_canta_values = [3, 2076, 8, 36/2076, 570/2076]

# Pie chartlar
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(mgv_canta_values[:3], labels=labels[:3], autopct='%1.1f%%', startangle=140)
plt.title('MGV Çanta - Pie Chartlar')

# Bar chartlar
plt.subplot(1, 2, 2)
plt.bar(labels[3:], mgv_canta_values[3:])
plt.title('MGV Çanta - Bar Chart')
plt.ylabel('Rate')
plt.ylim(0, 1)

plt.tight_layout()
plt.show()
```
